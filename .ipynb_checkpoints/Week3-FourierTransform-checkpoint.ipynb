# 3. Fourier analysis and applications

<p> Most of us are likely familiar with the Fourier series introduced by Joseph Fourier, which states that any periodic function can be expressed by an infinite sum of sines and cosines of Fourier frequencies. <p>
    
$$x(t)= c_{0} + \sum_{m=1}^{\infty} [c_{m}cos(2\pi(\dfrac{m}{N})t + s_{m}sin(2\pi(\dfrac{m}{N})t]$$ 

<p> Another way of looking at this is we're taking a function and breaking it down into all the frequency components present in the function. The Fourier transform is an extension of this Fourier series concept that is applied to signals that are aperiodic. Depending on whether the signal is continuous or discrete, and whether it is periodic or aperiodic, the exact terminology used in place of the Fourier transform / function varies. The table below summarizes this. </p>

![Fouriertable](FFTtable.png)

<p> For a continuous signal of length N (the signal extends from 0 to N seconds), the Fourier series yields an infinite number of sine and cosine waves whose frequencies will be 0/N Hz, 1/N Hz, 2/N Hz,....$\infty$/N Hz. The first Fourier frequency, $0/N = 0 Hz$, is called the <b> DC component </b> and is represented by $c_0$ in the very first equation.</p>
<br>

## <u> 3.1 Discrete Fourier Transform </u>
<p> Today, we will be focusing specifically on <b> Discrete Fourier Transform (DFT) </b>. DFT does not involve an infinite sum, but rather <b> includes only a limited range of frequencies </b>. This is because we only need a limited range of frequencies to represent discrete samples. Remember how adhering to the sampling theorem makes sure the sampled signal has no frequency greater than the Nyquist. So we don't need frequencies higher than the Nyquist to represent a discrete signal in the frequency space. The forward DFT (going from time domain to frequency domain) of a signal with N samples ($t = [0, 1, 2, ..., N-1]$) is given by, </p>

$$X_m=\dfrac{1}{N}\sum_{t=0}^{N-1}x(t)e^{(-2\pi jmt)/N}$$ 

<p> Note that for continuous signals, the above equation would be written using the integral instead of summation. The inverse DFT is used for transformation from the frequency domain back to the time domain. Say we know the Fourier frequency components $X_m$ and want to figure out what $N$ discrete samples of a signal $x(t)$ are, we use inverse DFT to do this. 
    
$$x(t) = \sum_{m=(-N/2)+1}^{N/2}X_m e^{(-2\pi jmt)/N}$$

<p>Run the code cells below to plot the Fourier spectrum and look at the Fourier frequencies of a real and complex sinusoid. </p>

```python
# Run this before any other cells
import numpy as np
import matplotlib.pyplot as plt
from sigUtil import *
import scipy.signal as signal
```

```python
## real valued sinusoidal signal
fs_sine = 100
t_sine = arangeT(0,1,fs_sine)
s_sine = np.sin(2*np.pi*5*t_sine)
S_sine = np.fft.fft(s_sine)
S_sine_freq = np.fft.fftfreq(len(t_sine), d=1/fs_sine)
print("Fourier frequencies = ", S_sine_freq)

fig = plt.figure(figsize=(7,5))
plt.plot(S_sine_freq, np.abs(S_sine)/len(t_sine))
# plt.plot(S_sine_freq, np.arctan(S_sine_freq.imag/S_sine_freq.real))
plt.xlabel("Frequemcy")
plt.ylabel("Magnitude")
plt.show()
```

```python
## complex valued sinusoidal signal
fs_comp = 100
t_comp = arangeT(0,1,fs_comp)
s_comp = np.exp(1j* 2 * np.pi * 5* t_comp)
S_comp = np.fft.fft(s_comp)
S_comp_freq = np.fft.fftfreq(len(t_comp), d=1/fs_comp)
print("Fourier frequencies = ", S_sine_freq)

fig = plt.figure(figsize=(7,5))
plt.plot(S_comp_freq, np.abs(S_comp)/len(t_comp))
# plt.plot(S_comp_freq, np.arctan(S_comp_freq.imag/S_comp_freq.real))
plt.xlabel("Frequemcy")
plt.ylabel("Magnitude")
plt.show()
```

## <u> 3.2 A few things about DFT </u>
<ul>
    <li> <b> Naming convention</b>. If you had paid close attention to the variables name used in the code and the markdown cells, you might have noticed that some variables are in uppercase and some are in lowercase. Lower case letters are used for time domain variables; and upper case letters are used for frequency domain representation. For example, $X = FFT(x)$, and $x = IFFT(X)$. </li>
    <br>
    <li> <b>FFT</b>. We discussed Discrete Fourier Transform (DFT) and Inverse Discrete Fourier Transform (IDFT) at the beginning. But when we are using Python to perform these functions, we use <i> numpy.fft</i> and <i> numpy.ifft </i>. The <i>fft</i> here stands for Fast Fourier Transform, which is just an efficient algorithm for computing DFT. </li> 
    <br>
    <li> <b> DFT resolution</b>. Frequency resolution of the DFT is given by $f_{s}/N$, i.e. the sampling frequency divided by the number of samples. This values basically tells us how far apart the spectra lines in the frequency domain are spaced. Some of the applications of DFT that we will look at involves manipulating this spectral resolution. </li>
    <br>
    <li> <b> Frequency ordering</b>. When we take the DFT of an N-sample time series, we end up with an array of length N. In the very beginning, we saw that the first element of the array is the DC component (0 Hz). The other frequencies are usually arranges in a specific way (in Python and in MATLAB); this is important to know in order to do frequency domain operations. The figure below shows the ordering. If you go back and look at the Fourier frequencies printed out in the above code cell results, you will notice this ordering. </li>
</ul>

 <ol> <li> The first term in the DC value (grey box). </li>

## <u> 3.3 Fourier transform pairs </u>

<p> Every time domain function typically has a corresponding frequency domain representation. We already looked at what sinusoids and chirps look like in the frequency domain. A sine wave has a single frequency component; a chirp has a range of frequency components. A couple of other common waveforms are an impulse function and a rectangular pulse. Run the code cells below to see what their frequency domain representations look like. </p>


```python
## Impulse function
imp = signal.unit_impulse(100)
IMP = np.fft.fft(imp)
IMP_freq = np.fft.fftfreq(len(imp), 1)

## rectangular pulse
rect = np.zeros(300)
rect[145:155] = 1

RECT = np.fft.fft(rect)
RECT_freq = np.fft.fftfreq(len(rect), 1)

fig, ax = plt.subplots(2,2)
ax[0,0].plot(imp) 
ax[0,1].plot(IMP_freq, np.abs(IMP))
ax[1,0].plot(rect) 
ax[1,1].plot(RECT_freq, np.abs(RECT))

ax[0,0].set_title("Time domain")
ax[0,1].set_title("Frequency domain")

fig.set_size_inches(6,4)
plt.show()
```

## <u> 3.4 Fourier transform operations </u>

<p> Similar to how waveforms in time domain have equivalent representations in frequency domain, there are also mathematical operations time domain that have frequency domain equivalent. We will look at a couple of those today </p>
    
#### 1. Frequency shifting
<p> Multiplying a (time domain) signal by a complex exponent of frequency $f_{0}$ ($e^{2\pi jf_{0}t}$) is equivalent to shifting the signal's frequency by $f_{0}$ in the frequency domain. We already tried this out in complex signal mixing in the first notebook. </p>


#### 2. Time shifting
<p> Shifting a signal in time domain by $t_{0}$ is equivalent to multiplying by a complex exponent $e^{2\pi jft_{0}}$ in the frequency domain. You can try this out by taking FFT of a sinusoid, multiplying the DFT by some exponential signal, taking the inverse DFT, and comparing against the original signal. </p>

#### 3. Convolution 
<p> Convolving two signals in time domain and taking the DFT of the resulting signal gives the same result as multiplying the Fourier transforms of the original signal. Filtering in DSP is typically implemented via convolution. We will see in the next notebook how this convolution property of DFTs is leveraged to filter signals. </p>

### 3.5.2 Band-limited Interpolation
<p> The DFT can be used to interpolate a signal without introducing new frequency content to it. This differs from something like a linear interpolation, a cubic spline, etc. because those interpolation methods can introduce higher frequency content to the interpolated signal. The premise is simple - take the DFT of a signal, zero pad it on both the low and high frequency sides, and then take the inverse DFT. Remember that the resolution of the DFT is determined by the number of samples in the signal you put into it - more signal samples means higher resolution. The inverse DFT works the same way - the more frequencies there are in your vector the more signal samples are generated from it.</p>

<b> 3.5.2.1 Band-limited interpolation with even number of samples </b>

```python
## band-limited interpolation with even number of samples
## Create and plot a sinusoid
fs = 20
dt = 1.0/fs   ## time interval between samples; will get used later when making interpolated signal

# Generate a simple sine wave at 2 Hz
t = arangeT(0,1,fs)
s = np.sin(2*np.pi*2*t)

fig, ax = plt.subplots(3,1)
ax[0].plot(t, s, 'o')
ax[0].set_xlabel("Time (s)")
ax[0].set_ylabel("Amplitude")

# Take the FFT
S = np.fft.fft(s)

# Remember the frequency ordering
f = np.fft.fftfreq(len(s), d=dt)
print("FFT Frequencies", f)

# Add in 0 padding to the middle, since that is 
# where the nyqist (high) frequencies are
# In this case we are doubling the number of frequency
# samples, which will double the sampling frequency too
S = np.concatenate((S[:10], np.zeros(len(S)), S[10:]))

t_interp = np.arange(len(S))*dt/2
# t_interp = arangeT(0,1,2*fs)
s_interp = np.real(np.fft.ifft(S))*2 # Need to multiply by 2 because we doubled length of freq vector

# Plot both signals
ax[1].plot(t, s, 'o')                   # Original signal in blue circles
ax[1].plot(t_interp, s_interp,'x')      # Interpolated signal in orange crosses
ax[1].set_xlabel("Time (s)")
ax[1].set_ylabel("Amplitude")

## Plot the FFT
ax[2].plot(np.fft.fftfreq(len(s), d=dt/2), np.abs(np.fft.fft(s)), 'o')
ax[2].plot(np.fft.fftfreq(len(s_interp), d=dt/2), np.abs(np.fft.fft(s_interp)), 'x')
ax[2].set_xlabel("Frequency (Hz)")
ax[2].set_ylabel("Magnitude")

fig.set_size_inches(10, 10)
plt.show()
```

<b> Linear interpolation </b>

```python
# Try linear interpolation
t_linear = np.arange(len(s)*2)*dt/2
s_linear = np.interp(t_linear, t, s)

plt.figure()
plt.plot(t, s, 'o') 
plt.plot(t_linear, s_linear, 'x')

plt.figure()
plt.plot(np.fft.fftfreq(len(s_interp), d=dt/2),np.abs(np.fft.fft(s_interp)), 'x')
plt.plot(np.fft.fftfreq(len(s_linear), d=dt/2), np.abs(np.fft.fft(s_linear)), 'o')
plt.show()
```

<b> 3.5.2.2 Band-limited interpolation with odd number of samples </b>

```python
## band-limited interpolation with odd number of samples
## Create and plot a sinusoid
fs = 25
dt = 1.0/fs   ## time interval between samples; will get used later when making interpolated signal

# Generate a simple sine wave at 2 Hz
t = arangeT(0,1,fs)
s = np.sin(2*np.pi*2*t)

fig, ax = plt.subplots(3,1)
ax[0].plot(t, s, 'o')
ax[0].set_xlabel("Time (s)")
ax[0].set_ylabel("Amplitude")

# Take the FFT
S = np.fft.fft(s)

# Remember the frequency ordering
f = np.fft.fftfreq(len(s), d=dt)
print("FFT Frequencies", f)

# Add in 0 padding to the middle, since that is 
# where the nyqist (high) frequencies are
# In this case we are doubling the number of frequency
# samples, which will double the sampling frequency too
S = np.concatenate((S[:int(len(S)/2)+1], np.zeros(len(S)-1), S[int(len(S)/2):]))

t_interp = np.arange(len(S))*dt/2
# t_interp = arangeT(0,1,2*fs)
s_interp = np.real(np.fft.ifft(S))*2 # Need to multiply by 2 because we doubled length of freq vector

# # Plot both signals
ax[1].plot(t, s, 'o')                   # Original signal in blue circles
ax[1].plot(t_interp, s_interp,'x')      # Interpolated signal in orange crosses
ax[1].set_xlabel("Time (s)")
ax[1].set_ylabel("Amplitude")

# ## Plot the FFT
ax[2].plot(np.fft.fftfreq(len(s), d=dt/2), np.abs(np.fft.fft(s)), 'o')
ax[2].plot(np.fft.fftfreq(len(s_interp), d=dt/2), np.abs(np.fft.fft(s_interp)), 'x')
ax[2].set_xlabel("Frequency (Hz)")
ax[2].set_ylabel("Magnitude")

fig.set_size_inches(10, 10)
plt.show()
```
