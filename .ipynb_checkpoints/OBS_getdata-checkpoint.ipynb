```python
import numpy as np
import matplotlib.pyplot as plt
import obspy
import matplotlib.pyplot as plt
from scipy import signal as signal
```

```python
## setting client
from obspy.clients.fdsn import Client
client = Client("IRIS")
```

```python
## retrieve stations in network Z6
starttime = obspy.UTCDateTime("2018-07-10T00:00:00-08:00")
endtime = obspy.UTCDateTime("2018-09-16T00:00:00-08:00")

inventory = client.get_stations(network="Z6",starttime=starttime, endtime=endtime, 
                        minlatitude=19, maxlatitude=20,
                        minlongitude=-156, maxlongitude=-154.1090,
                        level="channel" )

print(inventory)
inventory.plot(projection="local")
```

```python
tstart = obspy.UTCDateTime("2018-08-06T06:00:00.000")
tend = obspy.UTCDateTime("2018-08-02T06:45:00.000")

st = client.get_waveforms("Z6", "KSFF", "", "ELZ", tstart, tstart+3600)

st.plot()  
```
