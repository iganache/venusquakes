```python
## retrieve stations



## retrieve stations Tian
starttime = obspy.UTCDateTime("2014-01-01T00:00:00-08:00")
endtime = obspy.UTCDateTime("2017-01-01T00:00:00-08:00")

inventory = client.get_stations(starttime=starttime, endtime=endtime, 
                        minlatitude=64, maxlatitude=65.4067,
                        minlongitude=-151.4133, maxlongitude=-147.1090,
                        channel="BH*,HH*", level="response" )

print(inventory)
inventory.plot(projection="local")
```

```python
t = obspy.UTCDateTime("2010-02-27T06:45:00.000")

st = client.get_waveforms("IU", "ANMO", "00", "LHZ", t, t + 60 * 60)

st.plot()  
```
